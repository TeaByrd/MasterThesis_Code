```python
#working with vcf. SNPs to Patient
def SNPstoPatient(file):
    vcf_reader = vcf.Reader(filename=file)
    SNPtoPatientdic=dict()
    p = re.compile('AET[0-9]{4}')
    for record in vcf_reader:
        ID=record.ID
        if ID not in SNPtoPatientdic and ID!=None:
            SNPtoPatientdic[ID]=[]
            for call in record.samples:
                string_call=str(call)
                result = p.search(string_call)
                SNPtoPatientdic[ID].append(result.group(0))
    return SNPtoPatientdic
```

```python
SNPs2Patient=SNPstoPatient("../Aetionomy/riskSNPs_only.recode.vcf")
```
