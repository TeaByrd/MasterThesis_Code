```python
import pandas as pd
```

```python
#redefining the function
def prs(snps,catalog,log=False):
    
    if len(snps)==0:
        return 0
    
    #compile search pattern for input snps of interest
    snp_pattern = '|'.join(r"{}".format(x) for x in snps)
    catalog_filtered = catalog[catalog["snp"].str.contains(snp_pattern,na=False)]
    #
    catalog_filtered_pd=catalog_filtered[catalog_filtered["phewas phenotype"].str.contains("Parkinson's disease")] #filter catalog on pd phenotype only
    #
    naive_prs=catalog_filtered_pd["odds-ratio"].sum() #calculation of prs by choice of the user
    #
    return naive_prs
```

```python
catalog=pd.read_csv("phewascatalog_full.csv",delimiter = ',')
snps_genes=pd.read_csv("Final New Mappings /snps_genes_concatenated.csv",delimiter=",")
snps_list=list(set(snps_genes.rsID.tolist()))
```

```python
a=0
for SNP in snps_list:
    score=prs([SNP],catalog)
    print (SNP,score)
    if score >0.0:
        a+=1
```

```python
genes=pd.read_csv('LordickData/neurommsig_pd_full_genes.csv')
MechanismToGene=genes.to_dict("list")
#
#quickly remove nan values from gene lists to enhance efficiency
def remove_float_from_list(the_list):
    return [value for value in the_list if type(value)!=float]
#
MechanismToGene_noNans={x:remove_float_from_list(y) for x,y in MechanismToGene.items()}
```

```python
snps_genes
```

```python
mmsig_genes=[y for x,y in MechanismToGene_noNans.items()]
mmsig_genes_flat=[item for sublist in mmsig_genes for item in sublist]
mmsig_genes_set=list(set(mmsig_genes_flat)) # unique gene set from mmsig
```

```python
#subgraph/mechanism genes and snps
subgraph_genes=snps_genes[snps_genes["GENCODE_name"].isin(mmsig_genes_set)]
print (subgraph_genes, len(subgraph_genes))#21
```

```python
for mechanism in MechanismToGene_noNans.keys():
    query_df=snps_genes.loc[snps_genes["GENCODE_name"].isin(MechanismToGene_noNans[mechanism])]
    query_snps=query_df.rsID.tolist()
    score=prs(query_snps,catalog)
    print (mechanism,score) #10 Mechanisms summed up
```
