```python
def read_genotypes():
    risk_genotypes=pd.read_csv("../Skripte_SNPsMapping/Neurosnps_genotypes.csv")  #"../Skripte_SNPsMapping/Neurosnps_genotypes.csv")
    risk_genotypes=risk_genotypes.rename(columns={"Unnamed: 0":"AETID"})
    risk_genotypes=risk_genotypes.set_index("AETID",inplace=False)
    risk_genotypes=risk_genotypes.T
    return risk_genotypes
```

```python
genotypes=read_genotypes()
genotypes.to_csv("../Genotype.matrix.csv")
```

```python
def split_string(string): 
  
    # Split the string based on space delimiter 
    list_string = string.split(' ') 
      
    return list_string 
  
def join_string(list_string): 
  
    # Join the string based on '-' delimiter 
    string = '_'.join(list_string) 
      
    return string 
```

```python
MechanismToGene_noNans
MechanismToGene_noNans_newkeys={join_string(split_string(v)):k for v,k in MechanismToGene_noNans.items()}
```
